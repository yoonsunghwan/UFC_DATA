## Stance
* Fill in missing values for stance with other
* Replace every other stances with other if it is not orthodox or southpaw

```python
df['stance_1'].value_counts()
```

```python
df['stance_2'].value_counts()
```

```python
df['stance_1'] = df['stance_1'].replace(['Switch','Open Stance','Sideways',np.nan],'Other')
df['stance_2'] = df['stance_2'].replace(['Switch','Open Stance','Sideways',np.nan],'Other')
```

## Reach
* Fill in the missing values for reach with the fighters height

```python
df['reach_1'].fillna(df['height_1'],inplace = True)
df['reach_2'].fillna(df['height_2'],inplace = True)
```

## Age
* Fill in the missing ages for fighters with the other fighters age
* If both fighters are missing fill it with the mode

```python
df['age_1'].fillna(df['age_2'],inplace =True)
df['age_2'].fillna(df['age_1'],inplace=True)
```

```python
df['age_1'].fillna(df['age_1'].mode(),inplace=True)
df['age_2'].fillna(df['age_2'].mode(),inplace=True)
```
