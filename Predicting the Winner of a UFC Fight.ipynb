## Stance

* If both fighters are missing fill it with the mode

```python
df['age_1'].fillna(df['age_1'].mode(),inplace=True)
df['age_2'].fillna(df['age_2'].mode(),inplace=True)
```

### Assumption of LOGISTIC REGRESSION 
> requires minimal multicollineararity  
> observations are independent  
> target variable is binary

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
```

```python
def diff_time_fighters(df,fighter1,fighter2,fight_date):
    
    """Returns the value of the last win subtracted from current fight date 
       Returns the value of the last lost subtracted from the current fight date
       The final value is the difference between the two fighters
       """
    diff_time = dict.fromkeys(['fighter_since_win','fighter_since_lose'])
    fighter_since_win = dict.fromkeys([fighter1,fighter2]) 
    fighter_since_lose = dict.fromkeys([fighter1,fighter2])
    fighter1_lose_win = df[df['fighter_name'] == fighter1].groupby('fighter_outcome')['Date'].max()
    fighter2_lose_win = df[df['fighter_name'] == fighter2].groupby('fighter_outcome')['Date'].max()
    try:
        fighter_since_lose[fighter1] =\
        datetime.strptime(fight_date, '%Y-%m-%d') - datetime.strptime(fighter1_lose_win.iloc[0], '%Y-%m-%d')
    except:
        fighter_since_lose[fighter1] =\
        timedelta(days=0)
    try:
        fighter_since_win[fighter1] =\
        datetime.strptime(fight_date, '%Y-%m-%d') - datetime.strptime(fighter1_lose_win.iloc[1], '%Y-%m-%d')
    except:
        fighter_since_win[fighter1] =\
        timedelta(days=0)
    try:
        fighter_since_lose[fighter2] =\
        datetime.strptime(fight_date, '%Y-%m-%d') - datetime.strptime(fighter2_lose_win.iloc[0], '%Y-%m-%d')
    except:
        fighter_since_lose[fighter2] =\
        timedelta(days=0)
    try:
        fighter_since_win[fighter2] =\
        datetime.strptime(fight_date, '%Y-%m-%d') - datetime.strptime(fighter2_lose_win.iloc[1], '%Y-%m-%d')
    except:
        fighter_since_win[fighter2] =\
        timedelta(days=0)
    
    fighter_since_win_diff = fighter_since_win[fighter1] - fighter_since_win[fighter2]
    fighter_since_lose_diff = fighter_since_lose[fighter1] - fighter_since_win[fighter2]

    diff_time['fighter_since_win'] = fighter_since_win_diff.days
    diff_time['fighter_since_lose'] = fighter_since_lose_diff.days
    
    return diff_time
```

```python
diff_time_fighters(df = new_df, fighter1 = 'Chan Sung Jung', fighter2= 'Brian Ortega', fight_date = '2020-10-17')
```

```python
def diff_stats_fighters(df,fighter1,fighter2):
    diff = {}
    for c in df.columns:
        if df[c].dtype == 'float':
            diff[c] = df[df['fighter_name'] == fighter1][c].iloc[-1]- df[df['fighter_name'] == fighter2][c].iloc[-1]
            
        else:
            pass
    return diff
```

```python
diff_stats_fighters(df=new_df,fighter1='Chan Sung Jung', fighter2= 'Brian Ortega')
```

```python
def get_stance(df,fighter1,fighter2):
    stance = dict.fromkeys(['stance_1','stance_2'])
    stance['stance_1'] = df[df['fighter_name'] == fighter1]['stance_'].iloc[0]
    stance['stance_2'] = df[df['fighter_name'] == fighter2]['stance_'].iloc[0]
    return stance
```

```python
get_stance(new_df,'Chan Sung Jung','Brian Ortega')
```

```python
def get_full_stats(df,fighter1,fighter2,fight_date):
    full_stats = {**diff_time_fighters(df,fighter1,fighter2,fight_date),\
                  **diff_stats_fighters(df,fighter1,fighter2),\
                  **get_stance(df,fighter1,fighter2)
                 }
    return pd.DataFrame([full_stats])
```
